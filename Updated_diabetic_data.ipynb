```python
filled_data.head()
```

```python
def categorical_features(dataset):
    categorical_col = dataset.select_dtypes(exclude=np.number).columns.tolist()
    return categorical_col

categorical_features(filled_data)
```

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
cat_col = categorical_features(filled_data)
for i in cat_col:
  filled_data[i] = labelencoder.fit_transform(filled_data[i])
filled_data.head()
```

```python
from sklearn.model_selection import train_test_split
X = filled_data.drop('Bld_HbA1c',1)
y = filled_data['Bld_HbA1c']

train_df, test_df, train_label, test_label = train_test_split(X, y, train_size = 0.9, random_state = 42)
```

```python
from sklearn.preprocessing import LabelEncoder,MinMaxScaler,StandardScaler
MinMax = MinMaxScaler()
train_df[train_df.columns] = MinMax.fit_transform(train_df[train_df.columns])
test_df[test_df.columns] = MinMax.transform(test_df[test_df.columns])
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

param_grid = {'max_depth': [1,2,3],'max_features': ['auto','sqrt','log2'],'min_samples_leaf': [1,2,3,4],'n_estimators': [50,100]}
rndmFrst = RandomForestRegressor()
grid_search = GridSearchCV(rndmFrst,param_grid =param_grid,cv=6,scoring='neg_mean_squared_error')
grid_search.fit(train_df,train_label)
print("Optimal parameter values : ",grid_search.best_params_)
print("*************************************************************************")
print('Evaluating the model on unseen test data')


pred = grid_search.predict(test_df)
rmse= np.sqrt(mean_squared_error(pred,test_label))
print("Random Forest after GridSearch RMSE : ",rmse)
```

```python
pred_df = pd.DataFrame(pred.round(1),columns=['RandomForest'],index=test_label.index)
pred_df.head()
```

```python
RF_pred_act_df = pd.concat([pred_df,test_label],1)
RF_pred_act_df.head()
```
